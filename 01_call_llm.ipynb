# AI Essentials with Python

This notebook demonstrates the simplest possible way to call a large language model (LLM) from Python.

We send a single question to a model running locally on your machine via [Ollama](https://ollama.com) and print the answer.

## Prerequisites

Before running this notebook you have to:

- Download Ollama from <https://ollama.com/download>
- Install and run Ollama
- Pull a model of your choice, e.g.:
  ```bash
  ollama pull llama3.2
  ```
- Change the `MODEL` variable to the model you have pulled locally

````python
MODEL = "llama3.2"
````

## Install the `ollama` Python package

Run the cell below to install the official Ollama client for Python.

````python
import subprocess, sys
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "ollama"])
````

## Call the model
We send a question to the AI and print the response text.

````python
import ollama

QUESTION = "How many r's are in the word strawberry?"

messages = [
    {"role": "user", "content": QUESTION}
]

response = ollama.chat(model=MODEL, messages=messages)

print(f"Question: {QUESTION}")
print(f"Answer ({MODEL}):")
print(response.message.content)
````
